```python
import pandas as pd
```

```python
url = "https://drive.google.com/file/d/1HhOnZXyXkagtgd5TfzM61HPRuuykJ5AN/view?usp=sharing"
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Indicator Name'].unique()
```

Which country has the lowest death caused by nutrition in 2019?

```python
df_death = df[(df['Indicator Name'] == 'Cause of death, by communicable diseases and maternal, prenatal and nutrition conditions (% of total)')]
```

```python
df_death =df_death.sort_values(by=['2019'], ascending=True).head(1)
```

```python
df_death[['Country Name', '2019']]
```

Which country has the highest rate of diabetes prevalence in 2021?

```python
df_diabetes = df[(df['Indicator Name'] == 'Diabetes prevalence (% of population ages 20 to 79)')]
```

```python
df_diabetes =df_diabetes.sort_values(by=['2021'], ascending= False).head(1)
```

```python
df_diabetes[['Country Name', '2021']]
```

How much is the working age population of Mali in 2020?

```python
df_Mali = df[(df['Indicator Name'] == 'Age dependency ratio (% of working-age population)')]
df_Mali = df_Mali[['Country Name', '2020']]
```

```python
df_Mali[df_Mali['Country Name'] == 'Mali']
```

How many Indonesian adults (age 15+) are living with HIV in 2020?

```python
df_Indo = df[(df['Indicator Name'] == 'Adults (ages 15+) living with HIV')]
```

```python
df_Indo = df_Indo[['Country Name', '2020']]
```

```python
df_Indo[df_Indo['Country Name'] == 'Indonesia']
```

Which country has the lowest rate of literacy in 2020?

```python
df_literacy = df[df['Indicator Name'] == 'Literacy rate, adult total (% of people ages 15 and above)']
```

```python
df_literacy = df_literacy.sort_values(by=['2020'], ascending=True).head(1)
```

```python
df_literacy[['Country Name', '2020']]
```
